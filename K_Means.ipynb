# K-Means Clustering

## Part 1 


Read data from the file "iris.csv". Create a scatter plot with feature 'petal_length' on the x axis and 'petal_width' on the y axis.

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
###
### YOUR CODE HERE
###
data = pd.read_csv('iris.csv')
data.plot(kind='scatter',x='petal_length',y='petal_width')
```

## Part 2 

For k = 1, 2, 3, 4, and 5:
1. Perform k-means clustering. 
2. Create a scatter plot with feature 'petal_length' on x axis and feature 'petal_width' on y axis. For each scatter plot:
    i. Color each cluster with a different color. 
    ii. Mark the centroids obtained after running the k-means algorithm as red stars.


### For k = 1

```python
###
### YOUR CODE HERE
###
kmeans = KMeans(n_clusters=1)
x= data.iloc[:,[0,1,2,3]].values
y_kmeans = kmeans.fit_predict(x)
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'orange', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids')
plt.legend()
```

## Part 3

```python
distortions = []
x = data.iloc[:,[0,1,2,3]].values
K = range(1,6)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(x)
    distortions.append(kmeanModel.inertia_)
    

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method optimal k')
plt.show()
```
